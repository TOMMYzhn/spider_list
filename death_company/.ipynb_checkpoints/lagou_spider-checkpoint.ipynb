```python
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
import time
import pymysql
import pandas as pd


broswer = webdriver.Chrome()
def save_data(data):
    db = pymysql.connect(host='localhost',user='root',password='pass4321',port=3306,db='position_data')
    cursor = db.cursor()
    tables = 'position_table'
    keys = ','.join(data.keys())
    values = ','.join(['%s']*len(data))
    sql = 'INSERT INTO {table}({keys}) VALUES({values})'.format(table=tables,keys=keys,values=values)
    try:
        if cursor.execute(sql,tuple(data.values())):
            print('save success!')
            db.commit()
    except:
        print('save failed!')
        db.rollback()
    db.close()
try:
    broswer.get('https://www.lagou.com/')
    choose_city = broswer.find_elements_by_class_name('tab')[1]
    print(choose_city.text)
    choose_city.click()
    input_zhiwei = input('请输入需要查询的岗位:')
    input_name = broswer.find_element_by_id('search_input')
    input_name.send_keys(input_zhiwei)
    wait = WebDriverWait(broswer,1)
    click_but = broswer.find_element_by_id('search_button')
    click_but.click()
    wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'con_list_item')))
    time.sleep(1)
    # 获取需要爬取岗位的总页数
    total_page = broswer.find_elements_by_class_name('pager_not_current')[-1].text
    print(total_page)
    page = 1
    position_data = pd.DataFrame(columns=['position_name','position_location','position_time','position_experience',
    'position_company','position_infos','position_tags','position_introdce'])
    while page<= int(total_page):
        infos = broswer.find_elements_by_class_name('con_list_item')
        print('开始爬取第{}页的岗位信息数据'.format(page))
        page = page + 1
        for info in infos:
            # 岗位名称
            position_name = info.find_element_by_class_name('position_link').find_element_by_tag_name('h3').text
            # 岗位地点
            position_location = info.find_element_by_class_name('position_link').find_element_by_tag_name('em').text
            # 岗位发布时间
            position_time = info.find_element_by_class_name('format-time').text
            # 岗位工资、工作经验及学历
            position_experience = info.find_element_by_class_name('li_b_l').text
            # 招聘公司
            position_company = info.find_element_by_class_name('company_name').find_element_by_tag_name('a').text
            # 公司基本情况
            position_infos = info.find_element_by_class_name('industry').text
            # 岗位标签
            position_tags = info.find_element_by_class_name('list_item_bot').find_element_by_class_name('li_b_l').text
            # 公司基本介绍
            position_introdce = info.find_element_by_class_name('list_item_bot').find_element_by_class_name('li_b_r').text

            # po_data = {'position_name':position_name,'position_location':position_location,'position_time':position_time,'position_experience':position_experience,
            # 'position_company':position_company,'position_infos':position_infos,'position_tags':position_tags,'position_introdce':position_introdce}
            
            # save_data(po_data)

            position_data=position_data.append({'position_name':position_name,'position_location':position_location,'position_time':position_time,'position_experience':position_experience,'position_company':position_company,'position_infos':position_infos,'position_tags':position_tags,'position_introdce':position_introdce},ignore_index=True)
            print(position_name,position_location,position_time,position_experience,position_company,position_infos,position_tags,position_introdce)
        position_data.to_csv('position_infos.csv',mode='a',encoding='utf-8-sig')
        print('\n')
        next_page = broswer.find_element_by_class_name('pager_next')
        next_page.click()
        time.sleep(1)
    position_data.to_csv('position_data.csv',mode='a',encoding='utf-8-sig')
    broswer.quit()
except:
    print('error')
```
